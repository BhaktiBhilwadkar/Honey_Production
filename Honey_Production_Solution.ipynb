### 3. Display the percentage distribution of the data in each year using the pie chart.

```python
# calculating the percentage distribution of the data according to each year

year_per = honey_data.groupby('year').size()/len(honey_data)*100
```

```python
# plotting the actual pie chart


plt.figure(figsize=(6,6))
plt.pie(year_per, labels=year_per.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Data per Year')
plt.show()
```

**Observations:-**

1. Pie chart shows us  a proper visual distribution of the data
2. As we can see that the distribution is almost equally distributed per year with a slight difference in between.

### 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
# plotting displot of column(price per lb)


plt.figure(figsize=(6,6))
sns.distplot(honey_data['priceperlb'], color='brown')
plt.title("Distribution of Price per lb using displot", color='red')
plt.xlabel(('Price per lb'),color='red')
plt.ylabel(('Density'),color='red')
plt.show()
```

**Observation:-**

1. Dist Plot is used to explain the distribution of the data.
2. The data points of Price per lb can be easily observed using the distplot.
3. The price per lb indicates the average price of honey per pound based on sales.
4. If we observe, we can inferre that with increase in price per lb the sales decreases.

### 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
#plotting the scatterplot for the given data


plt.figure(figsize=(6,6))
plt.scatter(honey_data['numcol'], honey_data['prodvalue'],color = "blue")
plt.xlabel(('Number of Colonies'),color = "red")
plt.ylabel(('Production Value'), color = "red")
plt.title('Relation between Number of Colonies and Production Value.')
plt.show()
```

**Observations:-**

1. From the Scatter plot presented above, we can see that the column numcol and prodval are having a direct relationship i.e there is slightly positive correlation.
2. This means, as the number of colonies increases, the production value of honey also increases.
3. So to conclude; an Increase in number of colonies is increase to the production of honey.

### 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
#plotting a boxplot for the given data


plt.figure(figsize=(8,8))
sns.boxplot(x='year', y='prodvalue', data=honey_data)
plt.title('Relation between Year and Prodvalue', color='red')
plt.xlabel(('Year'),color = 'red')
plt.ylabel(('Prodvalue'),color = 'red')
plt.show()
```

**Observations:-**

1. The box plot is used to show the summary of the data. 
2. In the above plot, we used the box plot to summarize production value per year. 
3. We can also observed that as the year passes the sales increases and the median sales is also increased. 
4. The range of the production has been increasing every year and the value lies between 0.1 and 0.5 from the initial year(1998) to the last year(2012)
5. As we can see there are too may outliers but year 2010 has the most of it and then 2012 and other in the decresing order as shown above.

### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
# Plotting a PairPlot to view the multiple views.
plt.figure(figsize=(6,6))
sns.pairplot(honey_data, vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year'], hue='year', corner=True)
```

**Observations:-**

1. The Pairplot shows us the distribution  of single variables and the relationship between two variables. It is also used to find the correlation between variables.
2. The pairplot presented above shows that there is a direct relation between numcol and totalprod and prodvalue.
3. Prodvalue and totalprod also has a direct relation asit is moving in linear pattern. 
4. The Features such as numcol,totalprod and prodvalue are positively skewed in nature, whereas yieldpercol is normally distributed.

### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
#Firstly creating a subset of the dataframe 

sub_data = honey_data[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']]
```

```python
#calculating correlation matrix

corr = sub_data.corr()
```

```python
#plotting the heatmap using the corr

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

**Observations:-**

1. Heatmap is used to display the correlation matrix of the features in a dataset. 
2. The above heatmap almost gives the same inferences as the pair plot. 
3. It shows us the negative correlation with respect to priceperlb. 
4. It can be viewed that increase in yieldpercol results in reduction of priceperlb. 
5. The heatmap is illustrating a reasonable level of negative correlation.
6. Whearas if we observe more than there is strong colleration in totalprod and numcol.
